#### Pandas란?
- 행과 열로 이루어진 표 형태로 데이터 처리와 분석에 특화된 라이브러리

#### Pandas 구조
- 1. Series : 1차원 데이터 집합(index + value)
- 2. DataFrame : 2차원 표 형태의 데이터 집합
- 3. 1차원의 Series들이 모여 2차원의 DataFrame이 됨



```python
import pandas as pd
```

#### DataFlame
- 딕셔너리 이용 DF 생성하기

```python
data = {'서울' : [9904312,9631482],
       '부산' : [3448737, 3393191],
       '인천' : [2890451, 2632035],
       '대구' : [2466052, 2431774]}

ind = ['2015', '2010']
```

```python
population_df1 = pd.DataFrame(data, index = ind)
population_df1
# Series와는 달리 key값이 컬럼명으로 들어감(Series에서는 key값이 index로 들어감)
```

- 리스트 이용 DF 이용하기

```python
data = [[9904312, 3448737, 2890451, 2466052],
        [9631482, 3393191, 2632035, 2431774]]
ind=['2015', '2010']
col = ['서울', '부산', '인천', '대구']
```

```python
population_df2 = pd.DataFrame(data, index=ind, columns = col)
population_df2
```

#### 정리
- DF를 생성하는 방법은 2가지(딕셔너리 활용, 리스트 활용)
- 딕셔너리 방법은 key값이 컬럼명, value값이 데이터(위에서 아래방향)로 들어가고
- 리스트 방식은 데이터가 설정한 위치 그대로 들어감

```python
population_df2 = population_df2.T
population_df2
```

- 인덱스가 다른 시리즈끼리 연산을 진행하면 해당 인덱스의 값은 연산 불가(NaN : Not a Number)

- DataFrame

```python
df1 = pd.DataFrame({'2015':[9904312, 3448737, 2890451, 2466052],
                    '2010':[9631482, 3393191, 2632035, 2431774]})
df1
```

```python
df2 = pd.DataFrame({'2015':[9904312, 3448737, 2890451, 2466052],
                    '2010':[9631482, 3393191, 2632035, 2431774],
                    '2005':[9762546, 3512547, 2517680, 2456016]},
                   index=['서울', '부산','인천','대구'])
df2
```

```python
#인덱스랑 컬럼명이 다르면 연산 불가
df2 - df1
```

```python
df1.index= ['서울', '부산', '인천', '대구']
df1
```

```python
# 인덱스가 같으면 같은 컬럼명의 데이터끼리만 연산
df2 - df1
```

### Pandas에서 데이터 접근을 위한 인덱싱, 슬라이싱
- 인덱스 번호로 접근
- 인덱스 명으로 접근
- loc, iloc인덱서를 활용한 접근
- 불리언 인덱싱을 활용한 접근

- Series 인덱싱

```python
score = pd.Series({'java':70, 'python':95, 'html/css' : 80, 'ml': 85})
score
```

```python
# 인덱스 번호로 접근
score[1]
```

```python
# 인덱스 명으로 접근
score['python']
```

```python
# 인덱스 명 자체에 접근하고 싶을 때
score.index[1]
```

```python
# python에서 ML까지 인덱스 번호로 접근
score[1:]

#시리즈를 슬라이싱하면 인덱스 명과 값이 같이 출력된다.
```

```python
score.index[1:]
```

```python
score.values[1:]
```

```python
# pythonqnxj ML까지 인덱스 명으로 접근
score['python':'ml']
#인덱스 명으로 접근할 때는 끝 인덱스 명까지 포함
```

```python
# read_csv : csv 파일을 불러와서 DF으로 출력시켜주는 함수
# encoding :컴퓨터가 사람이 사용하는 문자를 인식할 수 있게 해주는 과정
# euc-kr : 한글 문자 전용 인코딩 방식
# index_col : 인덱스 컬럼 설정
score_data = pd.read_csv('data/score.csv', encoding='euc-kr', index_col='과목')
score_data
```

- DF 열(컬럼) 인덱싱

```python
# []같은 컬럼에 접근
score_data['1반']
# 시리즈 형태로 출력
```

```python
# []를 한번 더 씌우면 DF로 출력됨
score_data[['1반']]
```

```python
# 두개 이상의 컬럼 인덱싱(DF 출력시에는 반드시 [] 두 개를 사용할 것 )
score_data[['1반', '3반']]
```

- DF 행 인덱싱

```python
# DF에선 행 인덱싱을 위해서 슬라이싱 문법을 적용해줘야 함
score_data[0:1]
```

```python
score_data['수학': '영어']
```

```python
score_data[0:3:2] #맨 마지막 2는 두 행씩 뛰어넘어서 슬라이싱하는 방법
```

#### 정리
- DF의 기본적인 열 접근은 컬럼명으로,
- 행 접근은 슬라이싱 문법으로 접근 함

### loc, iloc를 활용한  인덱싱, 슬라이싱
- 한번에 행과 열에 모두 접근 가능
- 기본적으로 행에 먼저 접근

```python
score_data
```

```python
score_data.loc['과학']
```

```python
# 4반의 과학 점수
# DF명.loc[행, 열]
score_data.loc['과학', '4반']
```

```python
# 각 반의 과학 점수
score_data.iloc[4]
```

```python
# 4반의 과학 점수
score_data.iloc[4, 3]
```

- 2반의 사회 점수 데이터 접근해보세요(loc,iloc 각각 한 번씩 사용해보세요)

```python
score_data.loc['사회', '2반']
```

```python
score_data.iloc[3, 1]
```

- 2반과, 3반의 국어, 사회 점수에 접근해보세요

```python
score_data.loc['국어':'사회', '2반':'3반']
```

```python
score_data.iloc[2:4, 1:3]
```

#### 불리언 인덱싱(Boolean indexing)
- numpy의 array와 동일하며 DF에서 특정한 조건에 맞는 데이터에만 접근하기 위한 인덱싱 방법

```python
score_data = score_data.T
score_data
```

```python
# 영어 과목의 점수가 75점 이상인 반들을 찾아보자
# step1
score_data[score_data['영어'] >= 75]
```

```python
# step2
score_data[score_data['영어'] >= 75]['영어']
```

```python
score_data[score_data['영어'] >= 75]['영어'][0]
```

#### Pandas 요용한 함수들

#### 정렬함수
- index 및 컬럼명 기준 정렬 :sort_index()
- 값 기준 정렬 : sort_values()

```python
score_data = score_data.T
```

```python
score_data
```

- index 및 컬럼명 기준 정렬

```python
# index 기준 정렬
#한글일 경우 가나다 순으로 정렬
score_data.sort_index()
```

```python
# 컬럼명 기준 정렬
# axis = 0 : 행 방향
# axis = 1 : 열 방향
# drop 함수와는 반대다
score_data.sort_index(axis = 1, ascending=False)
```

- value기준 정렬

```python
# DF의 컬럼이나 index가 여러 개일 경우 그 중 기준을 설정해줘야 함
score_data.sort_values(by ='3반')
```

```python
score_data.sort_values(by ='3반', ascending=False)
```

```python
score_data.sort_values(by ='사회', axis=1, ascending=False)
```

#### 정리
- axis = 0 (행 방향 = 위에서 아래로)일 경우는 기준(by)을 컬럼명으로 잡고
- axis = 1 (열 방향 = 좌에서 우로)일 경우는 기준(by)를 인덱스로 잡음

```python
#3반 기준으로 정렬, 그 중에 같은 값이있다면 1반에서 한번 더 정렬
score_data.sort_values(by = ['3반', '1반'])
```

```python
# Series는 1차원이라 기준을 설정할 필요 없음!
score_data['3반'].sort_values()
```
