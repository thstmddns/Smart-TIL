### 목표 설정
- 간단한 직선을 그어서 w(가중치)/b(편향)

```python
# 경고창 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# 필요한 라이브러리 import
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# 데이터 준비
X = range(1, 6)
y = [2, 4, 6, 8, 10]
```

```python
# 그래프 상으로 X데이터와 y 데이터 확인해보기
plt.plot(X, y, marker = 'o')
plt.grid()
plt.show()
```

```python
# 문제와 정답 데이터를 데이터 프레임화 시키기
X_train = pd.DataFrame(X)
y_train = pd.DataFrame(y)
```

```python
# 모델 객체 생성
linear_model = LinearRegression()
```

```python
# 모델 학습
linear_model.fit(X_train, y_train)
```

```python
linear_model.predict([[3]]) # 예상되는 예측 값은? 6이 나올 것이다
# 주의점 : 선형회귀 모델은 데이터를 2차원으로 넣어줘야 한다.
# 1차원으로 데이터를 넣어줬을 시 에러가 발생한다.
```

```python
# 가중치와 평향값을 확인해보자
# 가중치 w값을 알아보자 : coef_
print(linear_model.coef_)
```

```python
# 편향(절편) b값을 알아보자 : intercept_
print(linear_model.intercept_)
```
